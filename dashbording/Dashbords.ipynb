```python
import boto3
import pandas as pd
from io import BytesIO

# Function to fetch the latest CSV file from MinIO
def fetch_latest_csv_from_minio():
    s3 = boto3.client(
        's3',
        endpoint_url='http://localhost:9000',
        aws_access_key_id='minioadmin',
        aws_secret_access_key='minioadmin'
    )
    
    # List objects in the bucket
    response = s3.list_objects_v2(Bucket='ecommerce-data')
    csv_files = [obj['Key'] for obj in response.get('Contents', []) if obj['Key'].endswith('.csv')]
    
    if not csv_files:
        raise Exception("No CSV files found in MinIO bucket.")
    
    # Get the latest CSV file
    latest_csv = sorted(csv_files)[-1]
    print(f"Fetching latest CSV file: {latest_csv}")
    
    # Download the file
    file_obj = s3.get_object(Bucket='ecommerce-data', Key=latest_csv)
    csv_data = file_obj['Body'].read()
    
    # Load CSV into a Pandas DataFrame
    df = pd.read_csv(BytesIO(csv_data))
    return df

# Example usage
orders_df = fetch_latest_csv_from_minio()
```

```python
from kafka import KafkaConsumer, TopicPartition
import json
import pandas as pd
import time

def fetch_logs_from_kafka(topic='user-interactions', max_messages=1000, timeout_ms=5000):
    """
    Fetch logs from a Kafka topic and return them as a Pandas DataFrame.

    Args:
        topic (str): The Kafka topic to consume from.
        max_messages (int): Maximum number of messages to fetch.
        timeout_ms (int): Timeout in milliseconds for polling.

    Returns:
        pd.DataFrame: A DataFrame containing the fetched logs.
    """
    # Initialize the Kafka consumer
    start_time = time.time()
    consumer = KafkaConsumer(
        bootstrap_servers='localhost:29092',  # External listener
        auto_offset_reset='earliest',
        enable_auto_commit=False,
        group_id=None,
        value_deserializer=lambda x: json.loads(x.decode('utf-8'))
    )
    print(f"Consumer initialized in {time.time() - start_time:.2f} seconds")
    
    # Assign the topic explicitly
    partition = TopicPartition(topic, 0)  # Replace 0 with the desired partition
    consumer.assign([partition])
    print(f"Assigned partitions: {consumer.assignment()}")
    
    # Fetch messages
    logs = []
    timeout_occurred = False
    start_time = time.time()
    
    while len(logs) < max_messages:
        # Poll for messages with a timeout
        batch = consumer.poll(timeout_ms=timeout_ms, max_records=max_messages - len(logs))
        
        if not batch:
            print("No messages received within the timeout period.")
            timeout_occurred = True
            break
        
        # Process the fetched messages
        print(f"Received messages: {batch}")
        for tp, messages in batch.items():
            logs.extend([msg.value for msg in messages])
            if len(logs) >= max_messages:
                break
    
    if not logs:
        raise Exception("No logs found in Kafka topic.")
    
    print(f"Fetched {len(logs)} messages in {time.time() - start_time:.2f} seconds")
    
    # Convert logs to a Pandas DataFrame
    logs_df = pd.DataFrame(logs)
    return logs_df

# Example usage
if __name__ == "__main__":
    try:
        logs_df = fetch_logs_from_kafka(topic='user-interactions', max_messages=100)
        print(logs_df.head())
    except Exception as e:
        print(f"Error: {e}")
```

# minio analysing

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style("whitegrid")

# Plot order status distribution
plt.figure(figsize=(10, 6))
status_counts = orders_df['status'].value_counts()
sns.barplot(x=status_counts.index, y=status_counts.values, palette="viridis")
plt.title("Order Status Distribution", fontsize=16)
plt.xlabel("Status", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

```python
# Convert order_date to datetime
orders_df['order_date'] = pd.to_datetime(orders_df['order_date'])

# Group by date and count orders
orders_over_time = orders_df.groupby(orders_df['order_date'].dt.date).size()

# Plot total orders over time
plt.figure(figsize=(12, 6))
orders_over_time.plot(kind='line', marker='o', color='blue')
plt.title("Total Orders Over Time", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Number of Orders", fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Group by status and calculate average price
avg_order_value = orders_df.groupby('status')['price'].mean().sort_values(ascending=False)

# Plot average order value
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_order_value.index, y=avg_order_value.values, palette="coolwarm")
plt.title("Average Order Value by Status", fontsize=16)
plt.xlabel("Status", fontsize=12)
plt.ylabel("Average Price", fontsize=12)
plt.show()
```
